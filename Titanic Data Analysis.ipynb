```python
# defining funciton to sort through the sex to find data on children
# anybody below 16 years of age is treated as a chold in this dataset

```

```python
def male_female_child(passenger):
    age,sex = passenger
    if age<16:
        return('child')
    else:
        return(sex)
```

```python
# adding a column named person and applying funciton to Age and Sex columns to sort children and add the child entry in
# the new person column

titanic_df['person'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```

```python
sns.factorplot('Pclass',data=titanic_df,kind='count',hue='person')
```

```python
titanic_df.head()
```

```python
# defining function to check if Alone column has value or zero
def family_alone(number):
    if number > 0:
        return('with family')
    else:
        return('alone')
```

```python
# applying function to Alone column

titanic_df['Alone'] = titanic_df['Alone'].apply(family_alone)
```
